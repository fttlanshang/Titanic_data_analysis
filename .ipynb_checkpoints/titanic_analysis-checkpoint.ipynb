```python
# How many survived?  
total_persons = len(titanic_data)
total_survival = titanic_data['Survived'].sum()
total_survival_ratio = total_survival * 1.0 / total_persons
print total_persons, total_survival
total_survival_ratio
```

```python
# Write a function to compute the survival ratio for different grouped data
def compute_survived_ratio_for_given_group(group_index):
    grouped_data = titanic_data.groupby(group_index)
    print grouped_data.size()
    survival_ratio_by_group = grouped_data.sum()['Survived'] * 1.0 / grouped_data.size()
    return survival_ratio_by_group
```

```python
# And what's the ratio for men & women?
# How can I get the count in the group??? titanic_data.groupby('Sex').size()
survival_ratio_by_sex = compute_survived_ratio_for_given_group('Sex')
survival_ratio_by_sex
#female has much higher survival ratio
```

```python
#  What about the Pclass?
survival_ratio_by_Pclass = compute_survived_ratio_for_given_group('Pclass')
survival_ratio_by_Pclass
```

```python
# What about combine the sex and Pclass together?
# note here should use an array when multiple factors
%matplotlib inline
survival_ratio_by_sex_Pclass = compute_survived_ratio_for_given_group(['Sex', 'Pclass'])
print survival_ratio_by_sex_Pclass
survival_ratio_by_sex_Pclass.unstack('Sex').plot(kind='bar')
```

```python
# Do the embarked port has relationship? Maybe someone get off board.
# Start from Endland Southampton, and passed Cherbourg in France and Queenstown in Ireland. Heading towards New York.
survival_ratio_by_embarked = compute_survived_ratio_for_given_group('Embarked')
survival_ratio_by_embarked
# maybe it's just a coincidence! maybe I can calculate the avergae ticket price to see if any one get off board.
# from 891 of the 2224 passengers and crew on board the Titanic. Why not all data? not avaiable?
```

```python
# Draw the histogram showing the relationship between age and numbers survived
%matplotlib inline
total_age_histogram = valid_age_titanic_data['Age'].hist(bins=8)
total_age_histogram.set_xlabel('Age')
total_age_histogram.set_ylabel('Numbers of persons')
total_age_histogram.set_title('The age distribution on the Titanic')
```

```python
survived_age_histogram = valid_age_titanic_data[valid_age_titanic_data['Survived'] == 1]['Age'].hist(bins=8)
survived_age_histogram.set_xlabel('Age')
survived_age_histogram.set_ylabel('Numbers of persons')
survived_age_histogram.set_title('The survived age distribution on the Titanic')
```

```python
# compute the survived ratio distribution related to age! There should be a bar chart!! The yaxis should be survived ratio!
# not familiar with this function!!!!!!
survived =  valid_age_titanic_data.groupby(pd.cut(valid_age_titanic_data['Age'], np.arange(0, 90, 10))).sum()['Survived'] 
total_counts_inrange = valid_age_titanic_data.groupby(pd.cut(valid_age_titanic_data['Age'], np.arange(0, 90, 10))).size()
print survived, total_counts_inrange
survived_ratio = survived * 1.0 / total_counts_inrange
survived_ratio_scatter = survived_ratio.plot(kind='bar')
survived_ratio_scatter.set_ylabel('Survived ratio')
survived_ratio_scatter.set_title('Survived ratio distribution related to age')
# We can see that the age is less important!! nearly a uniform distribution!
# Except the range (0, 10), maybe that's because there is less children.
```

```python
# How many children survived?
def compute_age_related_survived_ratio(specific_age, less_than):
    if less_than:
        total_inrange = valid_age_titanic_data[(valid_age_titanic_data['Age'] <= specific_age)]
    else:
        total_inrange = valid_age_titanic_data[(valid_age_titanic_data['Age'] >= specific_age)]
    total_survived_inrange = total_inrange[total_inrange['Survived'] == 1]
    print len(total_inrange), len(total_survived_inrange)
    return len(total_survived_inrange) * 1.0 / len(total_inrange)
```
